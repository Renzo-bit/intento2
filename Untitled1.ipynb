```python
# prompt: create a script to plot the math problem of 3n+1

import matplotlib.pyplot as plt

def collatz(n):
  """Generates the Collatz sequence for a given number."""
  sequence = [n]
  while n != 1:
    if n % 2 == 0:
      n = n // 2
    else:
      n = 3 * n + 1
    sequence.append(n)
  return sequence

# Get input from the user
n = int(input("Enter a starting number: "))

# Generate the Collatz sequence
sequence = collatz(n)

# Plot the sequence
plt.plot(sequence)
plt.xlabel("Iteration")
plt.ylabel("Number")
plt.title("Collatz Sequence for n = " + str(n))
plt.show()
```
